```python
import os
import torch  # We check PyTorch since it's the standard for this project
import sys

print("--- DIAGNOSTIC START ---")

# 1. Environment Check
print(f"Running on: {sys.executable}")
print(f"Current Directory: {os.getcwd()}")

# 2. GPU Availability Check (Crucial for Training)
# If this returns False, you are running on the CPU!
print(f"\nGPU Status:")
cuda_available = torch.cuda.is_available()
print(f"  CUDA/GPU Available: {cuda_available}")
if cuda_available:
    print(f"  Device Name: {torch.cuda.get_device_name(0)}")
else:
    print("  ⚠️ WARNING: Running on CPU. Ensure you have set the Colab Runtime to GPU.")

# 3. Google Drive Access Check
drive_path = '/content/drive/MyDrive'
print(f"\nGoogle Drive Status:")
if os.path.exists(drive_path):
    print("  ✅ Drive is mounted successfully.")
    # Check if your project folder exists
    project_folder = os.path.join(drive_path, 'COMP9517_Group_Project')
    if os.path.exists(project_folder):
        print(f"  ✅ Project folder found at: {project_folder}")
    else:
        print(f"  ⚠️ Project folder NOT found at: {project_folder}")
else:
    print("  ⚠️ Google Drive is NOT mounted. Run the drive.mount('/content/drive') command.")

# 4. Dataset Access Check
data_path = '/content/content/Team_Dataset'
print(f"\nDataset Status:")
if os.path.exists(data_path):
    print("  ✅ Team_Dataset folder found!")
    # Let's peek at the structure
    train_dir = os.path.join(data_path, 'train')
    if os.path.exists(train_dir):
        classes = os.listdir(train_dir)
        print(f"  ✅ Data check: Found {len(classes)} folders in the training set.")
    else:
        print("  ⚠️ 'train' subfolder missing inside Team_Dataset.")
else:
    print(f"  ⚠️ Team_Dataset folder missing at {data_path}")

print("\n--- DIAGNOSTIC COMPLETE ---")
```
