```python
import os
import random
import shutil
import json

SOURCE_TRAIN_DIR = '/content/train_mini'
SOURCE_TEST_DIR = '/content/val'

DEST_DIR = '/content/Team_Dataset'

# Project requirements
NUM_CLASSES = 500
TRAIN_SPLIT = 40
VAL_SPLIT = 10

all_classes = [d for d in os.listdir(SOURCE_TRAIN_DIR) if os.path.isdir(os.path.join(SOURCE_TRAIN_DIR, d))]


random.seed(42)
selected_classes = random.sample(all_classes, NUM_CLASSES)

print(f"Selected {len(selected_classes)} classes. Beginning data split...")

for split in ['train', 'val', 'test']:
    os.makedirs(os.path.join(DEST_DIR, split), exist_ok=True)
    for cls in selected_classes:
        os.makedirs(os.path.join(DEST_DIR, split, cls), exist_ok=True)

for cls in selected_classes:


    cls_train_path = os.path.join(SOURCE_TRAIN_DIR, cls)
    images = os.listdir(cls_train_path)


    random.shuffle(images)

    train_images = images[:TRAIN_SPLIT]           # First 40
    val_images = images[TRAIN_SPLIT:TRAIN_SPLIT + VAL_SPLIT] # Remaining 10

    for img in train_images:
        src = os.path.join(cls_train_path, img)
        dst = os.path.join(DEST_DIR, 'train', cls, img)
        shutil.copy2(src, dst)

    for img in val_images:
        src = os.path.join(cls_train_path, img)
        dst = os.path.join(DEST_DIR, 'val', cls, img)
        shutil.copy2(src, dst)

    cls_test_path = os.path.join(SOURCE_TEST_DIR, cls)

    if os.path.exists(cls_test_path):
        test_images = os.listdir(cls_test_path)
        for img in test_images:
            src = os.path.join(cls_test_path, img)
            dst = os.path.join(DEST_DIR, 'test', cls, img)
            shutil.copy2(src, dst)


mapping_path = os.path.join(DEST_DIR, 'selected_classes.json')
with open(mapping_path, 'w') as f:
    json.dump(selected_classes, f, indent=4)

print(f"Success! Dataset built at: {DEST_DIR}")
print(f"Class mapping saved to: {mapping_path}")
```
